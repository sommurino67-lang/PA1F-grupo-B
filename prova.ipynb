```python
num1 = float(input("Digite o primeiro número: "))
num2 = float(input("Digite o segundo número: "))
operacao = input("Digite a operação (+, -, *, /): ")

if operacao == '+':
  resultado = num1 + num2
  print(f"O resultado é: {resultado}")
elif operacao == '-':
  resultado = num1 - num2
  print(f"O resultado é: {resultado}")
elif operacao == '*':
  resultado = num1 * num2
  print(f"O resultado é: {resultado}")
elif operacao == '/':
  if num2 != 0:
    resultado = num1 / num2
    print(f"O resultado é: {resultado}")
  else:
    print("Erro: Divisão por zero não é permitida.")
else:
  print("Operação inválida.")
```

```python
valor_compra = float(input("Informe o valor da compra: "))

if valor_compra >= 100:
    desconto = valor_compra * 0.10
    total = valor_compra - desconto
    print(f"Desconto de 10% aplicado (R$ {desconto:.2f}). Total: R$ {total:.2f}")
elif valor_compra >= 50:
    desconto = valor_compra * 0.05
    total = valor_compra - desconto
    print(f"Desconto de 5% aplicado (R$ {desconto:.2f}). Total: R$ {total:.2f}")
else:
    print(f"Sem desconto. Total: R$ {valor_compra:.2f}")
```

```python
# Operador aritiméticos

# peidir ao usuários para digitar 02 números

# Entrada

a = int (input("digite um número:"))
b = int (input ("digite um número"))

#processamento

#soma

soma = a + b
subtracao = a - b
multiplicacao = a * b
divisao = a / b
div_int = a // b
resto = a % b
potencia = a ** b
# saída

print ("")
print ("-------------OPERADORES ARITIMÉTICOS-------------")
print("------------------------SOMA---------------------")
print("")
print(f"O resultado da soma de {a} mais {b} é igual  a : {soma}")
print("")
print("------------------------SUBITRAÇÃO---------------------")
print("")
print(f"O resultado da DIFERENÇA de {a} MENOS {b} é igual  a : {subtracao}")
print("")
print("---------------------------------------------")
print("------------------------MULTIPLICACAO------------=---------")
print("")
print(f"O resultado da MULTIPLICACAO de {a} VEZES {b} é igual  a : {multiplicacao}")
print("")
print("---------------------------------------------")
print("------------------------DIVISAO------------=---------")
print("")
print(f"O resultado da DIVISAO de {a} DIVIDIDO {b} é igual  a : {divisao}")
print("")
print("---------------------------------------------")
print(f"O resultado da DIVISAO de {a} DIVIDIDO {b} é igual  a : {div_int}")
print("------------------------DIVISAO------------=---------")
print("")
print(f"O resultado da DIVISAO de {a} DIVIDIDO por {b} o resto é  igual : {resto}")
print("")
print("---------------------------------------------")
print(f"O resultado da DIVISAO de {a} DIVIDIDO {b} é igual  a : {div_int}")
print("------------------------POTENCIA------------=---------")
print("")
print(f"O resultado da POTENCIA de {a} ELEVADO por {b} o resto é  igual : {potencia}")
print("")
print("---------------------------------------------")
```
